#Загрузка и преобразование данных

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import make_pipeline  # используем пайплайны для удобства
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate,GridSearchCV,RandomizedSearchCV
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
rs=42
```

```python
data=pd.read_csv("./heart.csv")
```

```python
data
```

```python
data.info()
```

```python
for name in data.keys().values.tolist():
  print(name,":")
  print(data[name].unique())
#print(data['Sex'].unique())
#print(data['ChestPainType'].unique())
```

```python
#Преобразуем данные при помощи LabelEncoder и OneHotEncoder
y = data['HeartDisease']
X = data.copy()
del X['HeartDisease']
Xp=pd.get_dummies( X, columns = [ 'ST_Slope','RestingECG','ChestPainType' ] )
le1 = LabelEncoder()
Xp.ExerciseAngina = le1.fit_transform(Xp.ExerciseAngina)
le2 = LabelEncoder()
Xp.Sex=le2.fit_transform(Xp.Sex)
Xp
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xp, y, test_size=0.2,random_state=rs)
```

```python
y
```

#Обучение модели

```python
score_params=['accuracy','recall','precision','f1']
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
cross_validate(model,X_train,y_train, cv=10, scoring=score_params)
```

# Оптимизация параметров

```python
params=dict(max_iter=[500,1000,2000,5000],
                                      penalty=['l1','l2','elasticnet'],
                                             solver=['lbfgs', 'liblinear', 'newton-cg']
                                      )
grid=GridSearchCV( LogisticRegression(random_state=rs),params,
                   n_jobs=None,
                   cv=10, scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
cross_validate(grid,X_train,y_train, cv=10, scoring=score_params)
```

```python
print(grid.best_score_)

print(grid.best_params_)

print(grid.best_estimator_)
```

```python
rcv=RandomizedSearchCV(LogisticRegression(random_state=rs),params,cv=10,verbose=1,n_iter=30,scoring='accuracy')
rcv.fit(X_train,y_train)
```

```python
cross_validate(rcv,X_train,y_train, cv=10, scoring=score_params)
```

```python
print(rcv.best_score_)

print(rcv.best_params_)

print(rcv.best_estimator_)
```

#Использование различных моделей

```python
from sklearn.svm import SVC
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
#Подготоваливаем словарь моделей
algos = [
  SVC(random_state=rs),
  DecisionTreeClassifier(random_state=rs),
  RandomForestClassifier(random_state=rs)
]
a_params= [
    dict(kernel=['linear', 'poly', 'rbf', 'sigmoid'],max_iter=[500,1000,2000,3000]),
    dict(min_samples_leaf=[1,5,10,30],max_depth=[5,10,20]),
    dict(min_samples_leaf=[1,5,10,30],max_depth=[5,10,20])
]
```

```python
i=0
rr=[None,None,None]
for algo in algos:
  print(a_params[i])
  rr[i]=RandomizedSearchCV(algo,a_params[i],cv=10,verbose=1,n_iter=30,scoring='accuracy')
  rr[i].fit(X_train,y_train)
  print(cross_validate(rr[i],X_train,y_train, cv=10, scoring=score_params))
  print(rr[i].best_score_)
  print(rr[i].best_params_)
  print(rr[i].best_estimator_)
  i+=1
```
